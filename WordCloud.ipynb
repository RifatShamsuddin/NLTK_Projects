```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

# Read text from a .txt file
with open('Example.txt', 'r', encoding='utf-8') as file:
    text = file.read()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import re

# Define a function to remove be-verbs, pronouns, adverbs, and prepositions
def filter_text(text):
    # Define lists of words to be removed
    be_verbs = ["am", "is", "are", "was", "were"]
    pronouns = ["i", "you", "he", "she", "it", "we", "they"]
    adverbs = ["very", "really", "quite"]
    prepositions = ["on", "in", "at", "with", "by", "for", "of", "about"]
    articles = ["the","a","an"]
    conjunctions = ["and", "but", "or", "nor", "for", "so", "yet", "although", "because", "since", "unless", "until", "while", "after", "before", "as", "if", "whether", "even though", "provided that"]


    # Tokenize the text into words
    words = re.findall(r'\b\w+\b', text.lower())

    # Filter out unwanted words
    filtered_words = [word for word in words if word not in be_verbs + pronouns + adverbs + prepositions + articles+ conjunctions]

    # Join the filtered words back into a string
    filtered_text = ' '.join(filtered_words)

    return filtered_text

# Filter the text
filtered_text = filter_text(text)

# Tokenize and count word frequencies
words = filtered_text.lower().split()
word_freq = Counter(words)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
